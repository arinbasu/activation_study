```julia
using Statistics
mean(skipmissing(kps.W_Epoch))
mean(skipmissing(bis.W_Epoch))
# summary_bis = [mean(skipmissing(bis.W_Epoch), median(skipmissing(bis.W_Epoch))]
mean_bis = mean(skipmissing(bis.W_Epoch))
med_bis = median(skipmissing(bis.W_Epoch))
min_bis = minimum(skipmissing(bis.W_Epoch))
max_bis = maximum(skipmissing(bis.W_Epoch))
alltogether = [mean_bis, med_bis, min_bis, max_bis]

function collate_stats(x)
    mean_x = mean(skipmissing(x))
    median_x = median(skipmissing(x))
    min_x = minimum(skipmissing(x))
    max_x = maximum(skipmissing(x))
    return([mean_x, median_x, min_x, max_x])
end

for_bis = collate_stats(bis.W_Epoch)
for_kps = collate_stats(kps.W_Epoch)
collate_stats(bis."W_Number of Epochs")
collate_stats(kps."W_Number of Epochs")
bis_hepochs = collate_stats(bis.H_Epoch)
kps_hepochs = collate_stats(kps.H_Epoch)
kp_n_epochs = collate_stats(kps."H_Number of Epochs")
bis_n_epochs = collate_stats(bis."H_Number of Epochs")
bis_axis_counts = collate_stats(bis."W_Axis 1 Counts")
```
